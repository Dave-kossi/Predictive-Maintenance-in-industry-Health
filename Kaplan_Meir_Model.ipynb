```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Lecture du Dataset
df = pd.read_csv("Predictive_Table.csv")

# Configuration pour l'analyse en jours

df['time_days'] = df['time']  # Explicitement renommer pour clarté

# 1. Analyse globale
kmf_overall = KaplanMeierFitter()
kmf_overall.fit(df['time_days'], df['event'], label='Toutes machines')
kmf_overall.plot_survival_function()
plt.title('Courbe de survie - Toutes machines')
plt.ylabel('Probabilité de fonctionnement')
plt.xlabel('Temps (jours)')
plt.grid(True)
plt.show()

print(f"Mediane de survie: {kmf_overall.median_survival_time_} jours")
print(f"Probabilité à 30 jours: {kmf_overall.survival_function_at_times(30).iloc[0]:.3f}")
print(f"Probabilité à 180 jours: {kmf_overall.survival_function_at_times(180).iloc[0]:.3f}")
```
